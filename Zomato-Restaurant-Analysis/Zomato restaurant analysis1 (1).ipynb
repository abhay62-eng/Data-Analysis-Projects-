# By using the subplot

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12,7))
# top cities
plt.subplot(1,2,1)
df['city'].value_counts().head(5).plot(kind="bar",color="orange")
plt.title("Top city",fontsize=18)
plt.xlabel('city')
plt.ylabel("count")
plt.show()


# Rating 
plt.figure(figsize=(12,5))
plt.subplot(1,2,2)
plt.hist(df['aggregate rating'],color='red')
plt.title('Rating Distribution')
plt.xlabel("Rating")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

df['cuisines'].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Top 5 Cuisines")
plt.ylabel("")  # remove y-label
plt.show()
```

# Scattered Plot to show the realtion ship between two variable

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df['average cost for two'],
    df['aggregate rating'],
    alpha=0.5
)
plt.xlabel("cost for two")
plt.ylabel("Rating")
plt.title("cost vs rating ")
plt.show()
```
