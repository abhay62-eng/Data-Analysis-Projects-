# Heatmap - missing pattern Visualization

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.title("Missing Values - Yellow = Missing")
plt.show()
```

# Bar Chart Of The Trageted Variable Analysis

```python
plt.figure(figsize=(4,4))
sns.countplot(x='Survived',data=df,palette=['#e74c3c','#2ecc71'],legend=False)
plt.xticks([0,1],['Died','Survived'])
plt.title('Titanic-survival Distribution')
plt.ylabel("Count")
plt.show()
```

# Age distribution

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
sns.histplot(df['Age'].dropna(),bins=30,kde=True,color='blue')
plt.title("Age Distribution")
plt.subplot(1,2,2)
sns.histplot(df['Fare'],bins=40,kde=True,color='#58a6ff')
plt.title("Fare Distribution (right skewed")
plt.tight_layout()
plt.show()

# Box plot - showing outlier 
plt.figure(figsize=(10,4))
sns.boxplot(data=df[['Age','Fare']],palette='pastel')
plt.title("Outlier Check - Box Plot")
plt.show()
```

# Categorical columns analysis, sex, pclass,embarked 

```python
plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
sns.countplot(data=df,x='Sex',hue='Survived',palette=['#e74c3c','#2ecc71'])
plt.title("sex vs survival")

# Survival by Pclass
plt.subplot(1,3,2)
sns.countplot(data=df,x='Pclass',hue='Survived',palette=['#e74c3c','#2ecc71'])
plt.title("Pclass vs Survival")

# Survival by Embarked
plt.subplot(1,3,3)
sns.countplot(data=df,x='Embarked',hue='Survived',palette=['#e74c3c','#2ecc71'])
plt.title("embarked vs survival")
plt.tight_layout()
plt.show()
```

# Correlation and heatmap


```python
numeric_df=df[['Survived','Pclass','Age','SibSp','Parch','Fare']]
# correlation matrix
corr=numeric_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,
           annot=True,
           fmt='.2f',
           cmap='coolwarm',
           center=0,
           square=True,
           linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

# pairplot 

```python
sns.pairplot(
    df[['Survived','Pclass',
        'SibSp','Parch','Fare']].dropna(),
    hue='Survived',
    palette={0:"#e74c3c",1:"#2ecc71"},
    diag_kind='kde',
    plot_kws={'alpha':0.6},
    height=2.5
)
plt.suptitle("Titanic Pairplot",y=1.02,fontsize=14)
plt.show()
```
