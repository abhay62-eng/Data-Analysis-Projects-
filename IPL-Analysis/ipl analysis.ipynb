# Top 10 Batsmen

```python
top_batsmen=df.groupby("batter")['batter_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen
```

```python
top_bowlers = df.groupby('bowler')['bowler_wicket'].sum().sort_values(ascending=False).head(10)

top_bowlers
```

# Toss vs Match Winner

```python
matches=df[['match_id','toss_winner','match_won_by']].drop_duplicates()

# check toss winner == match winner
matches['toss_win_match_win']=matches['toss_winner']==matches["match_won_by"]

# percentage 
result=matches['toss_win_match_win'].mean()*100
print("how many time a toss winner team wins the matches",result,"%")
```

# Strike Rate 

```python
balls=df.groupby("batter").size()
runs=df.groupby('batter')['batter_runs'].sum()

strike_rate = (runs/balls)*100
print(strike_rate.sort_values(ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt 
# Batsmen graph
top_batsmen.plot(kind='bar')
plt.title("Top 10 batsmen")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.show()

top_bowlers.plot(kind='bar')
plt.title("Top 10 bowler")
plt.xlabel("player")
plt.ylabel("wickets")
plt.show()
```

# PIE CHART OF TEAMS RUN

```python
import pandas as pd
import matplotlib.pyplot as plt 
teams_run=df.groupby("batting_team")['batter_runs'].sum()
plt.figure()
teams_run.plot(kind='pie',autopct='%1.1f%%',figsize=(8,8))
plt.title("Team_wise totel Run in IPL")
plt.show()
```

```python
# Top teams pie chart
top_teams=teams_run.sort_values(ascending=False).head(8)
top_teams.plot(
    kind='pie',autopct='%1.1f%%'
)
```

```python
# Season wise runs of each teams
seasons_wise=df.groupby(['season','batting_team'])['batter_runs'].sum().reset_index()
seasons_wise
```

```python
season_pivot=seasons_wise.pivot(index="season",columns="batting_team",values="batter_runs")
season_pivot.head()
```

```python
# Line chart Visualization 
top_teams = df.groupby('batting_team')['batter_runs'].sum().sort_values(ascending=False).head(5).index

filtered = season_pivot[top_teams]

filtered.plot()

plt.title("Top 5 Teams - Season-wise Runs")
plt.xlabel("Season")
plt.ylabel("Runs")

plt.xticks(rotation=45)
plt.show()
```
